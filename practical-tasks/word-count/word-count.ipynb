## word-count

## 1. Прочитать текстовый файл

```python
file_path = './texts/shakespeare/shakespeare.sonets.short.txt'
file = open(file_path)
text = file.read()
text = text.lower()
```

```python
text
```

## 2. Разбить длинную строку на слова.


```python
words = text.split(' ')
words = [w for w in words if w] # Отбросить пустые строки
words
```

## 3. Инициализировать словарь где ключом будет слово из текст и значением - количество повторений в тексте.


```python
word_count = {}
```

## 4. Перебрать все слова и по мере прохождения обновлять количество повторений в словаре.


```python
for w in words:
    if w not in word_count:
        word_count[w] = 0 ## встретили данное слово в первый раз
    word_count[w] += 1
```

## 5.  Выведем результат


```python
word_count
```

## 6. Чтоб было красивее - результат нужно отсортировать

```python
sorted(word_count.items(), key=lambda word_and_count: word_and_count[1], reverse=True)[:100]
```
